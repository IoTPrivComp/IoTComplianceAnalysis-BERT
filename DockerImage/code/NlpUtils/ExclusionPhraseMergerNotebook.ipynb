```python
from __future__ import unicode_literals
import spacy
from spacy.matcher import Matcher
```

```python
def mergeExcludePhrases(doc, vocab):
	def getLemma(ephrase):
		return u' '.join([ tok.lemma_ for tok in ephrase if tok.lemma_ != u':' ])

	results = []
	##########################################
	def callback(matcher, doc, i, matches):
		for match_id, start, end in matches:
			span = doc[start:end]
			results.append(span)
	##########################################
	matcher = Matcher(vocab)
	matcher.add(1,
		[[	{spacy.attrs.LOWER: "other"},
			{spacy.attrs.LOWER: "than"},
		]],on_match=callback)
	matcher.add(2,
		[[	{spacy.attrs.LOWER: "aside"},
			{spacy.attrs.LOWER: "from"},
		]],on_match=callback)
	matcher.add(3,
		[[	{spacy.attrs.LOWER: "apart"},
			{spacy.attrs.LOWER: "from"},
		]],on_match=callback)
	matcher.add(4,
		[[	{spacy.attrs.LOWER: "with"},
			{spacy.attrs.LOWER: "the"},
			{spacy.attrs.LOWER: "exception"},
			{spacy.attrs.LOWER: "of"}
		]],on_match=callback)
	matcher.add(5,
		[[	{spacy.attrs.LOWER: "except"},
			{spacy.attrs.ORTH: ":", 'OP' : '?'},
			{spacy.attrs.LOWER: "for"},
		]],on_match=callback)
	matcher.add(6,
		[[	{spacy.attrs.LOWER: "except"},
			{spacy.attrs.ORTH: ":", 'OP' : '?'},
			{spacy.attrs.LOWER: "when"},
		]],on_match=callback)
	matcher.add(7,
		[[	{spacy.attrs.LOWER: "except"},
			{spacy.attrs.ORTH: ":", 'OP' : '?'},
			{spacy.attrs.LOWER: "in"},
		]],on_match=callback)
	matcher.add(8,
		[[	{spacy.attrs.LOWER: "except"},
			{spacy.attrs.ORTH: ":", 'OP' : '?'},
			{spacy.attrs.LOWER: "to"},
		]],on_match=callback)
	matcher.add(9,
		[[	{spacy.attrs.LOWER: "except"},
			{spacy.attrs.ORTH: ":", 'OP' : '?'},
			{spacy.attrs.LOWER: "as"},
		]],on_match=callback)
	matcher.add(10,
		[[	{spacy.attrs.LOWER: "except"},
			{spacy.attrs.ORTH: ":", 'OP' : '?'},
			{spacy.attrs.LOWER: "under"},
		]],on_match=callback)
	matcher.add(11,
		[[	{spacy.attrs.LOWER: "except"},
			{spacy.attrs.ORTH: ":", 'OP' : '?'},
			{spacy.attrs.LOWER: "where"},
		]],on_match=callback)
	matcher.add(12,
		[[	{spacy.attrs.LOWER: "unless"},
			{spacy.attrs.ORTH: ":", 'OP' : '?'},
			{spacy.attrs.LOWER: "when"},
		]],on_match=callback)
	matcher.add(13,
		[[	{spacy.attrs.LOWER: "unless"},
			{spacy.attrs.ORTH: ":", 'OP' : '?'},
			{spacy.attrs.LOWER: "in"},
		]],on_match=callback)
	matcher.add(14,
		[[	{spacy.attrs.LOWER: "unless"},
			{spacy.attrs.ORTH: ":", 'OP' : '?'},
			{spacy.attrs.LOWER: "to"},
		]],on_match=callback)
	matcher.add(15,
		[[	{spacy.attrs.LOWER: "unless"},
			{spacy.attrs.ORTH: ":", 'OP' : '?'},
			{spacy.attrs.LOWER: "as"},
		]],on_match=callback)
	matcher.add(16,
		[[	{spacy.attrs.LOWER: "unless"},
			{spacy.attrs.ORTH: ":", 'OP' : '?'},
			{spacy.attrs.LOWER: "under"},
		]],on_match=callback)
	matcher.add(17,
		[[	{spacy.attrs.LOWER: "unless"},
			{spacy.attrs.ORTH: ":", 'OP' : '?'},
			{spacy.attrs.LOWER: "where"},
		]],on_match=callback)
	matcher.add(18,
		[[	{spacy.attrs.LOWER: "unless"},
			{spacy.attrs.ORTH: ":", 'OP' : '?'},
			{spacy.attrs.LOWER: "for"},
		]],on_match=callback)
	matcher.add(19,
		[[	{spacy.attrs.LOWER: "outside"},
			{spacy.attrs.LOWER: "of"},
		]],on_match=callback)



	matcher(doc)
	with doc.retokenize() as retokenizer:
		for ephrase in results:
			attrs = {"lemma":getLemma(ephrase)}
			retokenizer.merge(ephrase,attrs=attrs)
```
